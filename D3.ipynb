## Features
- number of exclaimation
- percent of uppercase
- contains smiley `:)`
- length of tweet
- percentage of punctuation
- contains negative words

Given that you've added the columns, use them as featurs for a NaivesBayesClassifier. `Sentiment` is target

```python
NEGATIVE_WORDS = ['damn', 'fuck']

df['conatins_negative_words'] = df['SentimentText'].str.contains('|'.join(NEGATIVE_WORDS))
```

```python
df['length_of_tweet'] =  df['SentimentText'].str.len()
```

```python
df['number of exclaimation'] =  df['SentimentText'].str.count("!")
```

```python
def pecentage_of_upper_case(str):
   count1=0
   count2=0
   for i in str:
         if(i.islower()):
               count1=count1+1
         elif(i.isupper()):
               count2=count2+1
   return count2 / len(str)
```

```python
df['percent_of_uppercase'] =  df['SentimentText'].apply(pecentage_of_upper_case)
```

```python
df.head()
```

```python
df['contains_smiley'] =  df['SentimentText'].str.find(":)") >= 0
```

```python
df.head()
```

```python
import string
def percentage_of_punctuation(str):
  count1=0
  for i in str:
         if i in string.punctuation:
            count1 = count1 + 1
         
  return count1 / len(str)

percentage_of_punctuation("a.2")
```

```python
df['percentage_of_punctuation'] =  df['SentimentText'].apply(percentage_of_punctuation)
```

```python
df.head()
```

```python
df.columns
```

```python
X = df[['conatins_negative_words', 'length_of_tweet', 'number of exclaimation', 'percent_of_uppercase', 'contains_smiley', 'percentage_of_punctuation']]
```

```python
y = df['Sentiment']
```

```python
X_train, X_test, y_train,y_test = train_test_split(X,y,test_size = .33)
```

```python
g = GaussianNB()
```
